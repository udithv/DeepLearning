```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
census = pd.read_csv('census_data.csv')
```

```
census.head()
```

```
census.columns
```

```
import tensorflow as tf
```

##Continous Features

```
#Continous Features
continous_features = ['age', 'education_num','capital_gain', 'capital_loss', 'hours_per_week']

age = tf.feature_column.numeric_column('age')
edu_num = tf.feature_column.numeric_column('education_num')
cap_gain = tf.feature_column.numeric_column('capital_gain')
cap_loss = tf.feature_column.numeric_column('capital_loss')
hpr = tf.feature_column.numeric_column('hours_per_week')
```

##Categorical Features

```
#Categorical Features
categorical_features = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'gender',  'native_country']

wclass = tf.feature_column.categorical_column_with_hash_bucket('workclass', hash_bucket_size=25)
edu = tf.feature_column.categorical_column_with_hash_bucket('education', hash_bucket_size=25)
marital_status = tf.feature_column.categorical_column_with_hash_bucket('marital_status', hash_bucket_size=25)
occup = tf.feature_column.categorical_column_with_hash_bucket('occupation', hash_bucket_size=25)
rela = tf.feature_column.categorical_column_with_hash_bucket('relationship', hash_bucket_size=25)
gender = tf.feature_column.categorical_column_with_vocabulary_list('gender', ['Female', 'Male'])
country = tf.feature_column.categorical_column_with_hash_bucket('native_country', hash_bucket_size=25)
race = tf.feature_column.categorical_column_with_hash_bucket('race', hash_bucket_size=25)
```

```
#Data Preprocessing

X = census.drop(['income_bracket'], axis=1)
Y = census['income_bracket']
```

```
X.head()
```

```
Y.head()
```

```
X[continous_features] = X[continous_features].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```
X.head()
```

```
Y.unique()
```

```
def label_fix(label):
    if label==' <=50K':
        return 0
    else:
        return 1
```

```
Y = Y.apply(label_fix)
```

##Train Test Split


```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```
feat_col = [age, edu_num, cap_gain, cap_loss, hpr, wclass, edu, marital_status, occup, rela, gender, country, race] 
```

```
len(feat_col)
```

```
batch_size=100
```

```
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=batch_size, num_epochs=1000, shuffle=True)
```

```
model = tf.estimator.LinearClassifier(feature_columns=feat_col, n_classes=2)
```

```
model.train(input_fn=input_func, steps=1000)
```

```
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=100, num_epochs=1, shuffle=False)
```

```
list(model.predict(input_fn=pred_input_func))
```

```
predictions = []

for pred in model.predict(input_fn=pred_input_func):
  predictions.append(pred['class_ids'].item(0))
```

```
y_pred = predictions
```

```
y_pred
```

```
from sklearn.metrics import classification_report
```

```
print(classification_report(y_test, y_pred))
```
